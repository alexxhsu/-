<h3>資料筆數</h3>

```python
aql='FOR doc IN StockInfo2 RETURN doc'
sub_collections=db.AQLQuery(aql, rawResults=False, batchSize=30000) #false結構化,資料子集合
print('筆數: %d' % len(sub_collections.response['result']))
```

<h3>此 SQL 資料庫表格需要衍生一個索引欄位，規則為：MOV0001，後四碼為序號</h3>

```python
for i in range(data.shape[0]):
    data['ID'][i]="MOV"+'%04d'%int(i+1)
data
```

```python
aql='FOR doc IN StockInfo2 update {_key: doc._key,"index" : "MOV"left(doc.timestamp,8)} into StockInfo2 RETURN doc'
sub_collections=db.AQLQuery(aql, rawResults=False, batchSize=30000) #false結構化,資料子集合
queryResult=sub_collections.response['result']
_resultSet=[]
for doc in queryResult:
    _resultSet.append(doc)
DataFrame(data=_resultSet)
```

<h3>將此資料集依照樣本時間 (timestamp) 由大到小排序</h3>

```python
aql='FOR doc IN StockInfo2 sort doc.`timestamp` desc RETURN doc'
sub_collections=db.AQLQuery(aql, rawResults=False, batchSize=30000) #false結構化,資料子集合
queryResult=sub_collections.response['result']
_resultSet=[]
for doc in queryResult:
    _resultSet.append(doc)
DataFrame(data=_resultSet)
```
