```python
import pandas
from glob import glob
```

<h1>取得資料檔案群</h1>

```python
data_folder = 'C:/Users/user/anaconda3/Projects/data/IBM-Stocks'
#檔案路徑設定

data_files = glob('%s/IBM-20??.csv' % data_folder)

data_files
```

<h3>載入資料檔</h3>

```python
dataframe_2017 = pandas.read_csv(data_files[0])
```

<h3>顯示部分資料</h3>

```python
dataframe_2017.head()
```

<h3>顯示資料欄位名稱</h3>

```python
dataframe_2017.columns
```

<h3>顯示欄位資料型態</h3>

```python
dataframe_2017.dtypes
```

<h3>變更欄位名稱</h3>

```python
dataframe_2017.rename(columns={
    'Date': 'StockDate', 
    'Open': 'StockOpen', 
    'High': 'StockHigh', 
    'Low': 'StockLow', 
    'Close': 'StockClose', 
    'Adj Close': 'StockAdjClose', 
    'Volume': 'StockVol'
}, inplace=True)

dataframe_2017.columns
```

<h3>增加新常數資料欄位</h3>

```python
dataframe_2017['StockID'] = 'IBM'
```

<h3>顯示資料欄位名稱</h3>

```python
dataframe_2017.tail(8)
```

<h3>計算欄位資料</h3>

```python
dataframe_2017['StockVol'] = dataframe_2017['StockVol'] / 1000.0
dataframe_2017.tail(8)
```

<h3>顯示欄位資料型態</h3>

```python
dataframe_2017.dtypes
```

<h3>指派資料索引鍵值</h3>

```python
dataframe_2018 = dataframe_2017.set_index(['StockID', 'StockDate'])
dataframe_2018
```

<h3>變更資料欄位順序</h3>

```python
new_columns = [
    'StockID', 'StockDate', 'StockOpen', 'StockHigh', 'StockLow', 'StockClose',
    'StockAdjClose', 'StockVol'
]

dataframe_2017[new_columns].head()
```

<h3>寫入資料檔</h3>

```python
dataframe_2017[new_columns].to_hdf('%s/IBM.h5' % data_folder, key='IBM2017', mode='w')
#hdf檔案(位址,檔名,模式)
```
