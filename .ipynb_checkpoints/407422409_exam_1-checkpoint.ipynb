<h3>建立表格</h3>

```python
_sql1 = [
    "CREATE TABLE IF NOT EXISTS",
    "MOVIE(NAME CHAR, YEAR INTEGER, DURATION INTEGER, GENRE CHAR, RATING CHAR, VOTES INTEGER, DIRECTOR CHAR, ACTOR1 CHAR, ACTOR2 CHAR, ACTOR3 CHAR, INDEX CHAR(7));"
]
sql1 = ' '.join(_sql1)
dbCursor.execute(sql1)
```

```python
_sql = 'TRUNCATE TABLE PUBLIC.MOVIE'
dbCursor.execute(_sql)
```

<h3>撰寫 Python 程式，逐筆產生 SQL INSERT 語句，將此 CSV 匯入該SQL 資料庫表格中</h3>

```python
_sql2=[
    "INSERT INTO",
    "MOVIE(NAME, YEAR,",
    "DURATION, GENRE, RATING,",
    "VOTES, DIRECTOR, ACTOR1, ACTOR2, ACTOR3, INDEX)",
    "VALUES('%s', %d, %d, '%s', %s, %d, '%s', '%s', '%s', '%s', '%s');"
] 
sql2 = ' '.join(_sql2)
for i in range(data.shape[0]):
    row = data.iloc[i]
    insert_sql = sql2 % (row[0].replace("'", "''"),
                         row[1],
                         row[2],
                         row[3],
                         row[4],
                         row[5],
                         row[6].replace("'", "''"),
                         row[7],
                         row[8].replace("'", "''"),
                         row[9].replace("'", "''"),
                         row[10])
    #print(insert_sql)
    dbCursor.execute(insert_sql)
```

<h3>7,8撰寫 SQL AGGREATION 語句，計算各 Genre 中之樣本數，記錄在 Jupyter Markdown 中</h3>

```python
_sql3 = [
    "SELECT Genre,COUNT(Genre) AS sum",
    "FROM MOVIE",
    "GROUP BY Genre"
]  

dbCursor.execute(' '.join(_sql3))
resultSet = dbCursor.fetchall()

rows = {
    'GENRE': [],
    'SUM': []
}

for row in resultSet:
    rows['GENRE'].append(row[0])
    rows['SUM'].append(row[1])

results = pandas.DataFrame(data=rows).set_index(['GENRE'])

results
```

<h3>11,12撰寫 SQL AGGREATION 語句，計算各 Rating 中之樣本數，記錄在 Jupyter Markdown 中</h3>

```python
_sql4 = [
    "SELECT Rating,COUNT(Rating) AS sum",
    "FROM MOVIE",
    "GROUP BY Rating"
]  

dbCursor.execute(' '.join(_sql4))
resultSet = dbCursor.fetchall()

rows = {
    'Rating': [],
    'SUM': []
}

for row in resultSet:
    rows['Rating'].append(row[0])
    rows['SUM'].append(row[1])

results = pandas.DataFrame(data=rows).set_index(['Rating'])

results
```
