```python
from pandasql import sqldf
pysqldf=lambda q:sqldf(q,globals())
```

```python
data_floder='D:/Projects/IBM-Stocks'
data_file=glob('%s/IBM-2018.csv'%data_floder)
data5=glob('%s/IBM-20??.csv'%data_floder)
data_file
data5
```

```python
data=pandas.read_csv(data_file[0])
```

```python
#第一題
data['Net'] = numpy.abs(data['Open']-data['Adj Close'])
data
```

```python
my_base_sql='SELECT * FROM data'
```

```python
net_avg=data['Net'].mean()
net_avg
```

```python
#第二題
my_sql=my_base_sql + ' WHERE Volume >5000000 AND Net >21.093216492000018 ;'
#net_avg=my_sql % str(data['Net'].mean())
pysqldf(my_sql)
```

```python
#第四題
my_sql=[
    'SELECT * FROM data A',
    'INNER JOIN',
    '(SELECT SUBSTR (Date,1,7) AS YM, MIN(Low) AS Low FROM data GROUP BY SUBSTR(Date,1,7)) B',
    'ON A.Low=B.Low'
]
pysqldf(' '.join(my_sql))
```

```python
#第五題
dataa=None
for i in range(0,4):
    d=pandas.read_csv(data5[i])
    my_sql=[
    'SELECT * FROM d A',
    'INNER JOIN',
     '(SELECT SUBSTR (Date,1,4) AS Y, MAX(Volume) AS Volume FROM d GROUP BY SUBSTR(Date,1,4)) B',
     'ON A.Volume=B.Volume',
    ]
    dataa1=pysqldf(' '.join(my_sql))
    dataa=pandas.concat([dataa,dataa1]) 
#dataa=dataa.sort_values(by='Y')
dataa['Stat'] = 'H'
dataa.set_index(['Y','Stat'])
dataa
```
