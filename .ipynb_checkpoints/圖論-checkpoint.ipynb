```python
import networkx as nx
import matplotlib.pyplot as plt
```

<h3>初始圖構</h3>

```python
G=nx.Graph()
```

<h3>建構圖資</h3>

```python
G.clear()
for i in range(5):
    G.add_node(i)
```

```python
G.add_edge(0,1) #0長出1
G.add_edge(0,2)
G.add_edge(1,3)
G.add_edge(1,5)
G.add_edge(2,4)
G.add_edge(2,5)
```

<h3>顯示圖</h3>

```python
fig,(ax1,ax2)=plt.subplots(dpi=240,ncols=2)
nx.draw(G,ax=ax1,with_labels=True,font_weight='bold')
nx.draw_shell(G,ax=ax2,with_labels=True,font_weight='bold')
fig.tight_layout()
plt.show()
```
