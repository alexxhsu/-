```python
db_ip='localhost'

db_conn=Connection(
    arangoURL='http://%s:8529'%db_ip,
    username='emprogria',
    password='emprogria'
)

db=db_conn['Emprogria']

db_collection = [
    db.collections['MRT_BR'],
    db.collections['MRT_R'],
    db.collections['MRT_G'],
    db.collections['MRT_O'],
    db.collections['MRT_BL'],
    db.collections['MRT_Y'],
    db.collections['node_info']
]
```

```python
def search_node_arango(conditions={'Name':'動物園'},numOfRows=4):
    global db
    aql_O='FOR doc IN MRT_O FILTER doc.StationName==@Name RETURN {"ID":doc._key,"Lane":doc.Lane,"PreStation":doc.PreStation,"NextStation":doc.NextStation,"Joint":doc.Joint,"Transit":doc.Transit,"Weight":doc.Weight}'
    aql_G='FOR doc IN MRT_G FILTER doc.StationName==@Name RETURN {"ID":doc._key,"Lane":doc.Lane,"PreStation":doc.PreStation,"NextStation":doc.NextStation,"Joint":doc.Joint,"Transit":doc.Transit,"Weight":doc.Weight}'
    aql_Y='FOR doc IN MRT_Y FILTER doc.StationName==@Name RETURN {"ID":doc._key,"Lane":doc.Lane,"PreStation":doc.PreStation,"NextStation":doc.NextStation,"Joint":doc.Joint,"Transit":doc.Transit,"Weight":doc.Weight}'
    aql_BL='FOR doc IN MRT_BL FILTER doc.StationName==@Name RETURN {"ID":doc._key,"Lane":doc.Lane,"PreStation":doc.PreStation,"NextStation":doc.NextStation,"Joint":doc.Joint,"Transit":doc.Transit,"Weight":doc.Weight}'
    aql_BR='FOR doc IN MRT_BR FILTER doc.StationName==@Name RETURN {"ID":doc._key,"Lane":doc.Lane,"PreStation":doc.PreStation,"NextStation":doc.NextStation,"Joint":doc.Joint,"Transit":doc.Transit,"Weight":doc.Weight}'
    aql_R='FOR doc IN MRT_R FILTER doc.StationName==@Name RETURN {"ID":doc._key,"Lane":doc.Lane,"PreStation":doc.PreStation,"NextStation":doc.NextStation,"Joint":doc.Joint,"Transit":doc.Transit,"Weight":doc.Weight}'
    
    sub_collections_O=db.AQLQuery(aql_O, rawResults=False, batchSize=numOfRows, bindVars=conditions)
    sub_collections_G=db.AQLQuery(aql_G, rawResults=False, batchSize=numOfRows, bindVars=conditions)
    sub_collections_Y=db.AQLQuery(aql_Y, rawResults=False, batchSize=numOfRows, bindVars=conditions)
    sub_collections_BL=db.AQLQuery(aql_BL, rawResults=False, batchSize=numOfRows, bindVars=conditions)
    sub_collections_BR=db.AQLQuery(aql_BR, rawResults=False, batchSize=numOfRows, bindVars=conditions)
    sub_collections_R=db.AQLQuery(aql_R, rawResults=False, batchSize=numOfRows, bindVars=conditions)
   
    queryResult_R=sub_collections_R.response['result']
    queryResult_O= sub_collections_O.response['result']
    queryResult_G=sub_collections_G.response['result']
    queryResult_BL= sub_collections_BL.response['result']
    queryResult_BR=sub_collections_BR.response['result']
    queryResult_Y= sub_collections_Y.response['result']


    return (queryResult_O,queryResult_R,queryResult_G,queryResult_BR,queryResult_Y,queryResult_BL)
```

<h3>使用  networkx 資料結構，列印兩站間北捷最短路徑各站資訊</h3>

```python
def find_shortest(source,target):
    shortest=single_source_dijkstra(G,source=source,target=target)
    return shortest
```
