```python
dataframe_2K = pandas.read_csv(data_files[my_file])
dataframe_2K
```

<h3>顯示部分資料</h3>

```python
dataframe_2K.tail()
```

<h3>查詢字串欄位包含某樣式資料</h3>

<h3>找資料庫最末筆</h3>

```python
my_last_month = dataframe_2K['Date'].sort_values().iloc[-1] #小到大,-1為倒數第一筆
my_last_month
```

```python
my_last_month = dataframe_2K['Date'].sort_values(ascending=False).iloc[0] #大到小
my_last_month #用最後一天找最後一個月
```

```python
my_last_month = int(my_last_month[5:7]) #5~6位
my_last_month 
```

```python
my_month = '-%02d-' % numpy.random.randint(1, my_last_month + 1) #1~12,兩位數不足補0
my_month
```

```python
my_monthly = dataframe_2K[dataframe_2K['Date'].str.contains(my_month)]#尋找包含某特徵的資料
my_monthly.head(10)
```

```python
my_condition = dataframe_2K['Date'].str.contains(my_month)
my_condition[:5]#真值表
my_condition[188:198]
```

```python
my_monthly = dataframe_2K[my_condition] #用[]把true的印出來
my_monthly.head(10)
```

<h3>計算數值資料欄位平均數及標準差</h3>

```python
my_monly_stats = (my_monthly['Open'].mean(), my_monthly['Open'].std())
my_monly_stats
```

```python
my_monly_stats2 = (my_monthly['Volume'].mean(), my_monthly['Volume'].std())
my_monly_stats2
```

<h3>比較數值資料欄位</h3>

<h3>上界條件查詢</h3>

```python
my_condition = my_monthly['Open'] < my_monly_stats[0] + 0.5 * my_monly_stats[1]
my_monthly_upper = my_monthly[my_condition]
my_monthly_upper.tail()
```

<h3>下界條件查詢</h3>

```python
my_condition = my_monthly['Open'] > my_monly_stats[0] - 0.5 * my_monly_stats[1]
my_monthly_lower = my_monthly[my_condition]
my_monthly_lower.head()
```

<h3>複合條件查詢</h3>

```python
my_monthly[
    (my_monthly['Open'] > my_monly_stats[0] - 0.5 * my_monly_stats[1]) &
    (my_monthly['Open'] < my_monly_stats[0] + 0.5 * my_monly_stats[1])
].set_index(['Date'])
```

```python
my_monthly2 = my_monthly[
    (my_monthly['Volume'] > my_monly_stats2[0] - 0.5 * my_monly_stats2[1]) &
    (my_monthly['Volume'] < my_monly_stats2[0] + 0.5 * my_monly_stats2[1])
]
my_monthly2.set_index(['Date'])
```

<h3>兩資料集交集合</h3>

```python
my_monthly_intersect = my_monthly_upper.merge(my_monthly_lower)
my_monthly_intersect.set_index(['Date'])
```

<h3>兩資料集差集合</h3>

```python
my_monthly[~my_monthly['Date'].isin(my_monthly_intersect['Date'])].set_index(['Date']) #不要(~)包含(isin)交集的資料=差集
```

<h3>兩資料集聯集合</h3>

```python
my_monthly_union = pandas.concat([my_monthly_intersect, my_monthly2]).drop_duplicates()#產生一個陣列來裝,並刪除重複的
my_monthly_union.set_index(['Date'])
```

<h3>篩選數值資料欄位最大值</h3>

```python
my_monly_max = None

for mon in range(1, my_last_month + 1):
    _month = '-%02d-' % mon #一個個月去看
    _df = dataframe_2K[dataframe_2K['Date'].str.contains(_month)] #某個月的所有資料 
    _max = _df[_df['Volume']==_df['Volume'].max()] #某個月的所有資料中的某一欄位最大值
    
    if my_monly_max is None:
        my_monly_max = _max #第一筆資料
    else:
        my_monly_max = pandas.concat([my_monly_max, _max])#合併每個月有最大值的資料
        
my_monly_max.sort_values(by='Date')
```

<h3>篩選衍生數值資料欄位最大值</h3>

```python
dataframe_2K['Net'] = numpy.abs(dataframe_2K['Adj Close'] - dataframe_2K['Open']) #絕對值

my_monly_max = None

for mon in range(1, 13):
    _month = '-%02d-' % mon
    _df = dataframe_2K[dataframe_2K['Date'].str.contains(_month)]
    _max = _df[_df['Net']==_df['Net'].max()]
    
    if my_monly_max is None:
        my_monly_max = _max
    else:
        my_monly_max = pandas.concat([my_monly_max, _max])
        
my_monly_max.sort_values(by='Date')
```
