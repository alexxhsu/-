```python
import networkx as nx
import matplotlib.pyplot as plt
```

<h3>初始圖構</h3>

```python
G=nx.Graph()
```

<h3>建構圖資</h3>

```python
G.clear()
for i in range(5):
    G.add_node(chr(65+i))
```

```python
G.add_edge(chr(65+0),chr(65+1)) #0長出1
G.add_edge(chr(65+0),chr(65+2))
G.add_edge(chr(65+1),chr(65+3))
G.add_edge(chr(65+1),chr(65+5))
G.add_edge(chr(65+2),chr(65+4))
G.add_edge(chr(65+2),chr(65+5))
```

<h3>顯示圖</h3>

```python
fig,(ax1,ax2)=plt.subplots(dpi=240,ncols=2)
nx.draw(G,ax=ax1,with_labels=True,font_weight='bold')
nx.draw_shell(G,ax=ax2,with_labels=True,font_weight='bold')
fig.tight_layout()
plt.show()
```

```python
node_names={
    'A':'明教',
    'B':'張三丰',
    'C':'溪',
    'D':'朱',
    'E':'一',
    'F':'诶府'
}
```

```python
for node in G.nodes:
    print(node,node_names[node])
```

```python
edge_names=dict()
for edge in G.edges:
    _start,_stop=edge
    _edge_name='%s-%s'%(node_names[_start],node_names)
    edge_names[edge]=_edge_name
    print(edge,_edge_name)
```
