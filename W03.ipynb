```python
from pandasql import sqldf
pysqldf=lambda q: sqldf(q,globals()) #建立函式，對q做sqldf(q,globals())的意思
```

```python
my_SQL=my_Base_SQL+' WHERE (Open BETWEEN %s AND %s) AND (Volume BETWEEN %s AND %s);'

pysqldf(my_SQL % (
    str(my_monly_stats[0]-0.5*my_monly_stats[1]),
    str(my_monly_stats[0]+0.5*my_monly_stats[1]),
    str(my_monly_stats2[0]-0.5*my_monly_stats2[1]),
    str(my_monly_stats2[0]+0.5*my_monly_stats2[1]))   
       ).set_index(['Date']).head()
```

```python
my_SQL=' SELECT SUBSTR (Date,1,7) AS YM,MAX(Open) AS Open FROM dataframe_2K GROUP BY SUBSTR(Date,1,7);'  #找每個月份open的最大值，設YM為欄位名稱
pysqldf(my_SQL).set_index(['YM'])
```

```python
my_SQL=[
    'SELECT * FROM dataframe_2K A',
    ' INNER JOIN(',
    'SELECT SUBSTR(Date,1,7) AS YM, MAX(Open) AS Open FROM dataframe_2K GROUP BY SUBSTR(Date,1,7)) B',
    'ON A.Open = B.Open'   
]  #方便閱讀，用array形式，把A設成dataframe_2K，B設dataframe_2K GROUP BY SUBSTR(Date,1,7))
pysqldf(' '.join(my_SQL)) #用空格做分隔
```

```python
my_SQL=[
    'SELECT * FROM dataframe_2K A',
    ' INNER JOIN(',
    'SELECT SUBSTR(Date,1,7) AS YM, MAX(Volume) AS Volume FROM dataframe_2K GROUP BY SUBSTR(Date,1,7)) B',
    'ON A.Volume = B.Volume'   
]
pysqldf(' '.join(my_SQL))
```

篩選衍生數值資料欄位最大值

```python
dataframe_2K['Net']=numpy.abs(dataframe_2K['Adj Close']-dataframe_2K['Open'])
#vectorization 減少計算複合型態資料的bigO時間
my_SQL=[
    'SELECT * FROM dataframe_2K A',
    ' INNER JOIN',
    '(SELECT SUBSTR(Date,1,7) AS YM, MAX(Net) AS Net FROM dataframe_2K GROUP BY SUBSTR(Date,1,7)) B',
    'ON A.Net = B.Net'   
]
pysqldf(' '.join(my_SQL))
```
