```python
import pandas
from glob import glob
import numpy
```

```python
data_floder='D:\Projects\IBM-Stocks'
data_files=glob('%s/IBM-2021.csv'%data_floder)
datao_files=glob('%s/ORCL (1).csv'%data_floder)
```

```python
dataframe_2021 = pandas.read_csv(data_files[0])
datao=pandas.read_csv(datao_files[0])
```

```python
dataframe_2021 
```

```python
datao
```

```python
datao['StockID'] = 'ORCL'
dataframe_2021 ['StockID'] = 'IBM'
```

```python
datao['Net'] = numpy.abs(datao['Open']-datao['Adj Close'])
datao['Scale'] = numpy.abs(datao['High']-datao['Low'])
dataframe_2021['Net'] = numpy.abs(dataframe_2021['Open']-dataframe_2021['Adj Close'])
dataframe_2021['Scale'] = numpy.abs(dataframe_2021['High']-dataframe_2021['Low'])

datao
```

```python
dataframe_2021
```

```python
datao_stats=(datao['Volume'].mean(),datao['Volume'].std())
datao_frame=datao[(datao['Volume']>datao_stats[0]-datao_stats[1])&
                   (datao['Volume']<datao_stats[0]+datao_stats[1])]
datao_frame
```

```python
dataframe_2021_stats=(dataframe_2021['Volume'].mean(),dataframe_2021['Volume'].std())
dataframe_2021_frame=dataframe_2021[(dataframe_2021['Volume']>dataframe_2021_stats[0]-dataframe_2021_stats[1])&
                   (dataframe_2021['Volume']<dataframe_2021_stats[0]+dataframe_2021_stats[1])]
dataframe_2021_frame
```

```python
datao_min=None
for m in range(1,10):
    month='-%02d-'%m
    df=datao_frame[datao_frame['Date'].str.contains(month)]
    minn=df[df['Scale']==df['Scale'].min()]
    if datao_min is None:
        datao_min=minn
    else:
        datao_min=pandas.concat([datao_min,minn])
df_o=datao_min.sort_values(by='Scale',ascending=False)
df_o
```

```python
dataframe_2021_min=None
for m in range(1,10):
    month='-%02d-'%m
    df=dataframe_2021_frame[dataframe_2021_frame['Date'].str.contains(month)]
    minn=df[df['Scale']==df['Scale'].min()]
    if dataframe_2021_min is None:
        dataframe_2021_min=minn
    else:
        dataframe_2021_min=pandas.concat([dataframe_2021_min,minn])
df_2021=dataframe_2021_min.sort_values(by='Scale',ascending=False)
df_2021
```

```python
union=pandas.concat([df_o,df_2021])
result=union.set_index(['StockID','Date'])
```

```python
result.rename(columns={
    'Open': '開盤價',  
    'High': '最高價', 
    'Low': '最低價', 
    'Close': '收盤價', 
    'Adj Close': '調整後收盤價', 
    'Volume': '成交量',
    'Net':'淨收入',
    'Scale':'規模'
}, inplace=True)
result
```
