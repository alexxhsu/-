```python
import numpy
from scipy.optimize import linprog
import matplotlib.pyplot as plt
```

<h3>求可生產最多產品數量</h3>
<ul>
    <li>生產要素: $x_1 \geq 0$ 及 $x_2 \geq 0$</li>
    <li>每生產一單位 z 產品，需要消耗: $5x_1$ 及 $4x_2$</li>
    <li>$x_2$ 數量 $\leq$ $60 - 4x_1$</li>
    <li>$2x_2$ 數量 $\leq$ $48 - 2x_1$</li>
</ul>

<table><tr><td>
$ 
max \\
\: \: \: \: z = 5x_1 + 4x_2 \\
s.t. \\
\: \: \: \: 4x_1 + x_2 \leq 60 \\
\: \: \: \: 2x_1 + 2x_2 \leq 48 \\
\: \: \: \: x_1 \geq 0 \\
\: \: \: \: x_2 \geq 0
$
    </td><td>
$ 
min \\
\: \: \: \: z^- = -5x_1 - 4x_2 \\
s.t. \\
\: \: \: \: 4x_1 + x_2 \leq 60 \\
\: \: \: \: x_1 + x_2 \leq 24 \\
\: \: \: \: x_1 \geq 0 \\
\: \: \: \: x_2 \geq 0
$
    <td></tr></table>

<h3>指派係數</h3>

```python
c = numpy.array([-5, -4])
A = numpy.array([[4, 1], [1, 1]])
b = numpy.array([60, 24])
```

<h3>求解</h3>

```python
res = linprog(
    c, A_ub=A, b_ub=b, 
    bounds=[(0, None), (0, None)],
    method='simplex'
)
```

<h3>最佳化</h3>

```python
res
```

<h3>最大化 $fun = -z$</h3>
轉換成正的

```python
if res.success:
    print(r'z = %f' % -int(res.fun))
```

<h3>圖解</h3> 
linspace 產生一萬個細微點 在0到40之間
ax.plot(x1, y1, label=r'$4x_1 + x_2 \leq 60$')
ax.plot(x1, y2, label=r'$2x_1 + 2x_2 \leq 48$') ＄號數學公式

patch 貼上去
'(%d, %d)' % (int(res.x[0]), int(res.x[1])) 顯示12,12
ax.legend 圖標


```python
fig, ax = plt.subplots(dpi=120)

x1 = numpy.linspace(0, 40, 10000)

y1 = 60 - 4*x1
y2 = 24 - x1

ax.plot(x1, y1, label=r'$4x_1 + x_2 \leq 60$')
ax.plot(x1, y2, label=r'$2x_1 + 2x_2 \leq 48$')

circle1 = plt.Circle((12.0, 12.0), 2.0, color='yellow')
ax.add_patch(circle1)

ax.text(res.x[0]+1, res.x[0]+1, '(%d, %d)' % (int(res.x[0]), int(res.x[1])), fontsize=8, color='green')

ax.fill_between(x1, 0, y1, where=(y1 <= 12), facecolor='gray')
ax.fill_between(x1, 0, y2, where=(y2 >= 12), facecolor='gray')

ax.set_xlim(0, 40)
ax.set_ylim(0, 40)

ax.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

ax.set_xlabel(r'$x_1$')
ax.set_ylabel(r'$x_2$')

plt.show()
```
