```python
import pandas
from glob import glob
```

```python
data_folder = '/Users/justin/Projects/DB/data/IBM-Stocks'

data_files = glob('%s/IBM-20??.csv' % data_folder)
data_files
```

```python
dataframe_2018=pandas.read_csv(data_files[0])
```

```python
dataframe_2018.head()
```

```python
dataframe_2018.columns
```

```python
dataframe_2018.dtypes
```

```python
dataframe_2018.rename(columns={
    'Date': 'StockDate', 
    'Open': 'StockOpen', 
    'High': 'StockHigh', 
    'Low': 'StockLow', 
    'Close': 'StockClose', 
    'Adj Close': 'StockAdjClose', 
    'Volume': 'StockVol'
}, inplace=True)

dataframe_2018.columns
```

```python
dataframe_2018['StockID'] = "IBM"
```

```python
dataframe_2018.head()
```

```python
dataframe_2018['StockVol'] = dataframe_2018['StockVol'] / 1000.0
```

```python
dataframe_2018.head()
```

```python
dataframe_2018.set_index(['StockID', 'StockDate'])
```

```python
new_columns = [
    'StockID', 'StockDate', 'StockOpen', 'StockHigh', 'StockLow', 'StockClose',
    'StockAdjClose', 'StockVol'
]

dataframe_2018[new_columns].head()
```

```python
dataframe_2018[new_columns].to_hdf('%s/IBM.h5' % data_folder, key='IBM2018', mode='w')
```
