```python
datazzz=(dataframe['Volume'].mean(),dataframe['Volume'].std())
```

```python
datazzzup=dataframe[dataframe['Volume']<datazzz[0]+datazzz[1]]
```

篩選出在欄位【Volume】平均數一個標準差內的數值資料，儲存為一個DataFrame

```python
my_monly_min = None

for mon in range(1, 10):
    _month = '-%02d-' % mon       
    _df = dataframe[dataframe['Date'].str.contains(_month)] # 各月份資料的df
    _min = _df[_df['Scale']==_df['Scale'].min()] 
    
    if my_monly_min is None:
        my_monly_min = _min 
    else:
        my_monly_min = pandas.concat([my_monly_min, _min]) 
        
anotherdf=my_monly_min.sort_values(by='Scale',ascending=False)
anotherdf
```
