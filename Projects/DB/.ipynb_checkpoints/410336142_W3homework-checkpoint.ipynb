```python
from pandasql import sqldf
pysqldf = lambda q : sqldf(q, globals())
```

產生衍生欄位NET，為OPEN減去ADJ CLOSE 的絕對值

```python
dataframe_2k["Net"] = numpy.abs(dataframe_2k["Adj Close"] - dataframe_2k['Open'])
dataframe_2k
```

透過本周上課的方式(pysqldf)，篩選出【Volume大於5000000與NET大於平均】交集的資料

```python
Netmean =(dataframe_2k['Net'].mean())
```

```python
my_Base_SQL ="SELECT * FROM dataframe_2k "
```

```python
my_SQL = my_Base_SQL + ' WHERE (Volume > %s) AND (Net > %s);'
pysqldf(my_SQL%(
        str(5000000),
        str(Netmean)
       ))
```

透過本周上課的方式(pysqldf)，找出各月LOW的最小值是哪一天，新增【YM】欄位儲存月份的值，使用INNER JOIN合併

```python
my_SQL='SELECT SUBSTR(Date, 1, 7) AS YM, MIN(LOW) AS LOW FROM dataframe_2k GROUP BY SUBSTR(Date, 1, 7);'
pysqldf(my_SQL).set_index(['YM'])
```

```python
my_SQL = [
    'SELECT * FROM dataframe_2k A',
    'INNER JOIN ',
    '(SELECT SUBSTR(Date, 1, 7) AS YM, MIN(LOW) AS LOW FROM dataframe_2K GROUP BY SUBSTR(Date, 1, 7)) B',
    'ON A.LOW = B.LOW'
]

 

my_SQL = ' '.join(my_SQL)
my_SQL
pysqldf(my_SQL)
```
