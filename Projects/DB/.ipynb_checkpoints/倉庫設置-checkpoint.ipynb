```python
import numpy
import pandas
```

```python
packages = []
num_of_pack = 8

for i in range(num_of_pack):
    packages.append(dict(id='PCK-%d' % (i + 1), 
                         volume=numpy.random.randint(1, 5) * 500, 
                         weight=numpy.random.randint(1, 5) * 300, 
                         piece=numpy.random.randint(1, 5) * 100))
    
package_dataset = pandas.DataFrame(data=packages)
package_dataset
```

```python
storages = []
num_of_bin = 64

for i in range(num_of_bin):
    storages.append(dict(id='STG-%02d' % (i + 1), 
                         max_volume=numpy.random.randint(50, 100) * 500, 
                         max_weight=numpy.random.randint(10, 80) * 300,
                         occupied_volume=0, occupied_weight=0,
                         packages=[])                         
                   )
    
storage_dataset = pandas.DataFrame(data=storages)
storage_dataset.head()
```

```python
num_of_shipment = 100

for i in range(num_of_shipment):
    package = package_dataset.sample(1)
    
    for j in range(storage_dataset.shape[0]):
        storage = storage_dataset.iloc[j]

        occupied_weight = storage['occupied_weight'] + package['weight']
        max_weight = storage['max_weight']
        
        if max_weight.item() >= occupied_weight.item():
            occupied_volume = storage['occupied_volume'] + package['volume']
            max_volume = storage['max_volume']
            
            if occupied_volume.item() <= max_volume.item():
                storage['packages'].append(package['id'].values[0])

storage_dataset.head()
```
