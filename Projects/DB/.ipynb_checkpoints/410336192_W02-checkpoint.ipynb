```python
import pandas
from glob import glob
import numpy
```

```python
dataframe_orcl= pandas.read_csv('C:/Projects/DB/data/ORCL.csv')
dataframe_orcl.head(10)
```

```python
dataframe_orcl['StockID'] = 'ORCL'
dataframe_orcl.head()
```

```python
dataframe_orcl['Net'] = numpy.abs(dataframe_orcl['Open'] - dataframe_orcl['Adj Close'])
dataframe_orcl['Scale'] = dataframe_orcl['High'] - dataframe_orcl['Low']
dataframe_orcl.head()
```

```python
stat=(dataframe_orcl['Volume'].mean(),dataframe_orcl['Volume'].std())
stat
```

```python
df_orcl_Volume=dataframe_orcl[
    (dataframe_orcl['Volume'] > stat[0] - stat[1] )&
    (dataframe_orcl['Volume'] < stat[0] + stat[1] )  
    ]
df_orcl_Volume.set_index(['Date'])
```

```python
my_last_month=dataframe_orcl['Date'].sort_values().iloc[-1] 
my_last_month=int(my_last_month[5:7])
my_last_month
```

```python
my_monly_min = None

for mon in range(1, my_last_month + 1):
    _month = '-%02d-' % mon       # 1=-01- 2=-02-....
    _df = dataframe_orcl[dataframe_orcl['Date'].str.contains(_month)] # 各月份資料的df
    _min = _df[_df['Scale']==_df['Scale'].min()] #各月分最大值(每月可能有重複的成交量)
    # _max是篩選資料的邏輯判斷
    
    if my_monly_min is None:
        my_monly_min = _min # 1月最大值先被丟入當my_monly_max
    else:
        my_monly_min = pandas.concat([my_monly_min, _min]) #每個月最大值聯集
        
df_orcl_min=my_monly_min.sort_values(by='Scale',ascending=False)
df_orcl_min
```

```python
dataframe_2021 = pandas.read_csv('C:/Projects/DB/data/IBM-2021.csv')
dataframe_2021.head(10)
```

```python
dataframe_2021['StockID'] = 'IBM'
dataframe_2021.head()
```

```python
dataframe_2021['Net'] = numpy.abs(dataframe_2021['Open'] - dataframe_2021['Adj Close'])
dataframe_2021['Scale'] = dataframe_2021['High'] - dataframe_2021['Low']
dataframe_2021.head()
```

```python
stat=(dataframe_2021['Volume'].mean(),dataframe_2021['Volume'].std())
stat
```

```python
df_2021_Volume=dataframe_2021[
    (dataframe_2021['Volume'] > stat[0] - stat[1] )&
    (dataframe_2021['Volume'] < stat[0] + stat[1] )  
    ]
df_2021_Volume.set_index(['Date'])
```

```python
my_last_month=dataframe_2021['Date'].sort_values().iloc[-1] 
my_last_month=int(my_last_month[5:7])
my_last_month
```

```python
my_monly_min = None

for mon in range(1, my_last_month + 1):
    _month = '-%02d-' % mon       # 1=-01- 2=-02-....
    _df = dataframe_2021[dataframe_2021['Date'].str.contains(_month)] # 各月份資料的df
    _min = _df[_df['Scale']==_df['Scale'].min()] #各月分最大值(每月可能有重複的成交量)
    # _max是篩選資料的邏輯判斷
    
    if my_monly_min is None:
        my_monly_min = _min # 1月最大值先被丟入當my_monly_max
    else:
        my_monly_min = pandas.concat([my_monly_min, _min]) #每個月最大值聯集
        
df_2021_min=my_monly_min.sort_values(by='Scale',ascending=False)
df_2021_min
```

```python
my_union=pandas.concat([df_orcl_min,df_2021_min]).drop_duplicates()
my_union.set_index(['StockID','Date']) 
```

```python
my_union.columns
```

```python
my_union.rename(columns={
    'Date': '日期',   
    'Open': '開盤價', 
    'High': '最高價', 
    'Low': '最低價', 
    'Close': '收盤價', 
    'Adj Close': '調整的收盤價', 
    'Volume': '成交量',
    'StockID': '股票名稱',
    'Net': '淨值',
    'Scale': '震盪'
}, inplace=True)

my_union.columns
```
