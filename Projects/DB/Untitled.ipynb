```python
from pandasql import sqldf
pysqldf = lambda q : sqldf(q, globals())
```

```python
my_SQL='SELECT SUBSTR(Date, 1, 4) AS YM, MAX(Open) AS Open FROM dataframe_2k GROUP BY SUBSTR(Date, 1, 4);'
pysqldf(my_SQL).set_index(['YM'])
```

新的欄位 YM 撈年份跟月份的資料 每個月份的Open各自最大值 顯現出MAX出來 要使用 GROUP BY YM的部分

```python
my_SQL = [
    'SELECT * FROM dataframe_2k A',
    'INNER JOIN (',
    'SELECT SUBSTR(Date, 1, 7) AS YM, MAX(Open) AS Open FROM dataframe_2K GROUP BY SUBSTR(Date, 1, 7)) B',
    'ON A.Open = B.Open'
]

 

my_SQL = ' '.join(my_SQL)
my_SQL
```

```python
pysqldf(my_SQL)
```
