```python
import networkx
```

```python
G=networkx.Graph()
```

```python
#節點
G.add_nodes_from(list(range(10))) #0~9
```

```python
#路徑
G.add_edge(0,1)
G.add_edge(1,2)
G.add_edge(1,3)
G.add_edge(2,5)
G.add_edge(3,6)
G.add_edge(4,7)
G.add_edge(4,8)
G.add_edge(4,9)
G.add_edge(9,5)
G.add_edge(9,6)
G.add_edge(0,4)
```

```python
#繪圖
pos=networkx.shell_layout(G)
```

```python
fig,ax=plt.subplots(figsize=(6,4),dpi=120)
networkx.draw(G,pos,with_labels=True,font_weight='bold',font_color='white',ax=ax)
plt.show()
```

```python
#尋找最短路徑(0到9)
shortest=networkx.shortest_path(G,source=0,target=9)
```

```python
'→'.join(map(str,shortest))
```

```python
#所有節點之最短路徑組合
routes={} 
```

```python
for _source in list(G.nodes): 
    print('節點:%s'%_source)
    
    for _target in [n for n in list (G.nodes) if _source !=n]:
        #不包括自己,0到1~9的距離
        shortest=networkx.shortest_path(
            G,source=_source,target=_target)#(source開始,target結束)
        _edge='→'.join(map(str,sorted(shortest))) #sorted排序方式
        
        if _edge in routes:
            routes[_edge]+=1
        else:
            routes[_edge]=1
        print('\t',_edge)
```

```python
for _edge in sorted(routes):
    _nodes=_edge.split('→')
    print(_nodes)
```

```python
for _source in list(G.nodes):
    print('節點:%s'%_source)
    for _target in [n for n in list(G.nodes) if _source!=n]:
        shortest=networkx.shortest_path(
        G,source=_source,target=_target)
       
        _edge='→'.join(map(str,sorted(shortest))) #sorted排序方式
        routes3.append(_edge)
        print('\t',_edge)
```

```python
def get_all_shortest_path(G):
    routes=set()
    for _source in list (G.nodes):
        for _target in [n for n in list(G.nodes) if _source !=n]:
            #不包括自己0到1~9
            shortest=networkx.shortest_path(
                G,source=_source,target=_target)#(source開始,target結束)
            _edge='→'.join(map(str,sorted(shortest))) #sorted排序方式
            routes.add(_edge)
    return routes
```

```python
all_shortest_path=get_all_shortest_path(G)
all_shortest_path
```

```python
for _edge in sorted (all_shortest_path):
    _nodes=_edge.split('→')
    print(_nodes)
```
