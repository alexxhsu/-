```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy
```

<h3>初始圖構</h3>

```python
G=nx.Graph()
```

<h3>建構圖資</h3>

```python
G.clear()
for i in range(5):
    G.add_node(chr(65+i),
              weight=numpy.random.randint(10,50))
```

```python
G.add_edge(chr(65+0),chr(65+1),weight=numpy.random.randint(10,50)) #0長出1
G.add_edge(chr(65+0),chr(65+2),weight=numpy.random.randint(10,50))
G.add_edge(chr(65+1),chr(65+3),weight=numpy.random.randint(10,50))
G.add_edge(chr(65+1),chr(65+5),weight=numpy.random.randint(10,50))
G.add_edge(chr(65+2),chr(65+4),weight=numpy.random.randint(10,50))
G.add_edge(chr(65+2),chr(65+5),weight=numpy.random.randint(10,50))
```

<h3>顯示圖</h3>

```python
fig,(ax1,ax2)=plt.subplots(dpi=240,ncols=2)
nx.draw(G,ax=ax1,with_labels=True,font_weight='bold')
nx.draw_shell(G,ax=ax2,with_labels=True,font_weight='bold')
fig.tight_layout()
plt.show()
```

```python
node_names={
    'A':'明教',
    'B':'張三丰',
    'C':'溪',
    'D':'朱',
    'E':'一',
    'F':'诶府'
}
for node in G.nodes(data=True):
    if'weight' in node[1]:
        print('%s\t%-8s\t%d'%(node[0],node_names[node[0]],node[1]['weight']))
    else: print('%s\t%-8s\t%d'%(node[0],node_names[node[0]],node[1]['weight']))
```

```python
edge_names=dict()
for edge in G.edges(data=True):
    _start,_stop=edge
    _edge_name='%s-%s'%(node_names[_start],node_names)
    edge_names[edge]=_edge_name
    if 'weight' in edge[2]:
        print('%s\t%-8s\t%d'%(edge[:2],_edge_name,edge[2]['weight'])
    else:
        print('%s\t%-8s\t%d'%(edge[:2],_edge_name,edge[2]['weight'])
```
